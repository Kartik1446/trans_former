# Micrograd — Intuition (Short)

* What is it?

Micrograd is a tiny library that helps you understand how **gradients (derivatives)** are computed automatically.



* Smart Numbers

```python
from micrograd.engine import Value
a= Value(-4.0)
b= Value(2.0)
c= a+b
d= a*b+ b**3
c += c+1
c+= 1+c+ (-a)
d+= d*2 + (b+a).relu()
d+= 3*d+(b-a).relu()
e= c-d
f= e**2
g= f/2.0
g+= 10.0/f
print(f'{g.data:.4f}')
g.backward()
print(f'{a.data:.4f}')
print(f'{b.data:.4f}')
```
